```python
import pandas as pd
import glob
import os
```

```python
files_list = glob.glob("data/*.csv")

funding_data = pd.concat(map(lambda file: pd.read_csv(file, encoding="cp1252"), files_list))
```

```python
funding_data = funding_data.apply(lambda x: x.astype(str).str.lower())
funding_data.columns = map(str.lower, funding_data.columns)
funding_data.columns = funding_data.columns.str.replace(' ', '_')
funding_data.columns = funding_data.columns.str.replace('(','')
funding_data.columns = funding_data.columns.str.replace(')','')
```
